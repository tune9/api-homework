## Part 1 - Personal Finance Planner

### Collect Crypto Prices Using the `requests` Library

```python
# Set current amount of crypto assets
# YOUR CODE HERE!
my_btc=1.2
my_eth=5.3
```

```python
# Crypto API URLs
btc_url = "https://api.alternative.me/v2/ticker/Bitcoin/?convert=CAD"
eth_url = "https://api.alternative.me/v2/ticker/Ethereum/?convert=CAD"
```

```python
# Fetch current BTC price
# YOUR CODE HERE!
btc_data = requests.get(btc_url)
btc_content = btc_data.json()

# Fetch current ETH price
# YOUR CODE HERE!
eth_data = requests.get(eth_url)
eth_content = eth_data.json()



btc_price = btc_content['data']['1']['quotes']['USD']['price']
eth_price = eth_content['data']['1027']['quotes']['USD']['price']
```

```python
# Compute current value of my crpto
# YOUR CODE HERE!
my_btc_value = btc_price*my_btc
my_eth_value = eth_price*my_eth 

# # Print current crypto wallet balance

print(f"The current value of your {my_btc} BTC is ${my_btc_value:0.2f}")
print(f"The current value of your {my_eth} ETH is ${my_eth_value:0.2f}")
```

### Collect Investments Data Using Alpaca: `SPY` (stocks) and `AGG` (bonds)

```python
# Set current amount of shares
my_agg = 200
my_spy = 50
```

```python
# Set Alpaca API key and secret
# YOUR CODE HERE!
api_key = os.getenv("ALPACA_API_KEY")
secret_key = os.getenv("ALPACA_SECRET_KEY")

# Create the Alpaca API object
# YOUR CODE HERE!
api = tradeapi.REST(
    api_key, 
    secret_key, 
    api_version="v2"
)
```

```python
type(secret_key)
```

```python
# Format current date as ISO format
# YOUR CODE HERE!


# Set the tickers
tickers = ["AGG", "SPY"]

# Set timeframe to '1D' for Alpaca API
timeframe = "1D"
today = pd.Timestamp("2022-03-17", tz='America/New_York').isoformat()
# Get current closing prices for SPY and AGG
# (use a limit=1000 parameter to call the most recent 1000 days of data)
# YOUR CODE HERE!

df_ticker = api.get_bars(
    tickers, 
    tradeapi.TimeFrame(1, tradeapi.TimeFrameUnit.Day),
    start=today,
    end=today,
    limit=1000, 
   ).df

    
# Preview DataFrame
# YOUR CODE HERE!
df_ticker.head()
```

```python
# Pick AGG and SPY close prices
# YOUR CODE HERE!
agg_close_price = df_ticker['close'][0]
spy_close_price = df_ticker['close'][1]
# Print AGG and SPY close prices
print(f"Current AGG closing price: ${agg_close_price}")
print(f"Current SPY closing price: ${spy_close_price}")
```

```python
# Compute the current value of shares
# YOUR CODE HERE!
my_spy_value = spy_close_price*my_spy
my_agg_value = agg_close_price*my_agg
# Print current value of shares
print(f"The current value of your {my_spy} SPY shares is ${my_spy_value:0.2f}")
print(f"The current value of your {my_agg} AGG shares is ${my_agg_value:0.2f}")
```

### Savings Health Analysis

```python
# Set monthly household income
# YOUR CODE HERE!
monthly_income=12000
# Consolidate financial assets data
# YOUR CODE HERE!
crypto = my_eth_value+my_btc_value
shares = my_spy_value+my_agg_value

# Create savings DataFrame
# YOUR CODE HERE!

data = [crypto, shares]
df_savings = pd.DataFrame(data, columns=['amount'])
df_savings.index = ['crypto', 'shares']
# Display savings DataFrame
display(df_savings)
```

```python
# Plot savings pie chart
# YOUR CODE HERE!
df_savings.plot(kind='pie', subplots=True, title='Composition of Personal Savings')
```

```python
# Set ideal emergency fund


emergency_fund = monthly_income * 3

# Calculate total amount of savings
# YOUR CODE HERE!
savings= crypto+shares

# Validate saving health
# YOUR CODE HERE!
if savings>emergency_fund:
    print("Congratulations! You have enough money in your savings account.")
    
elif savings==emergency_fund:
    print("Congratulations on reaching your financial goal! keep it up.")

else:
    print(f"You are ${emergency_fund-savings} away from meeting your financial goal!")
    
```

## Part 2 - Retirement Planning

### Monte Carlo Simulation

```python
# Set start and end dates of five years back from today.
# Sample results may vary from the solution based on the time frame chosen
# start_date = pd.Timestamp('2016-05-01', tz='America/New_York').isoformat()
# end_date = pd.Timestamp('2021-05-01', tz='America/New_York').isoformat()
```

```python
# Get 5 years' worth of historical data for SPY and AGG
# (use a limit=1000 parameter to call the most recent 1000 days of data)
# YOUR CODE HERE!
df_agg_stock_data = api.get_bars(['AGG'], tradeapi.TimeFrame(1, tradeapi.TimeFrameUnit.Day), "2016-05-01", "2021-05-01", limit=1000, adjustment='raw').df
df_agg_stock_data.drop(columns=['vwap', 'symbol', 'trade_count'], inplace=True)
df_agg_stock_data.columns = pd.MultiIndex.from_product([['AGG'], df_agg_stock_data.columns])



df_spy_stock_data = api.get_bars(['SPY'], tradeapi.TimeFrame(1, tradeapi.TimeFrameUnit.Day), "2016-05-01", "2021-05-01", limit=1000, adjustment='raw').df
df_spy_stock_data.drop(columns=['vwap', 'symbol', 'trade_count'], inplace=True)
df_spy_stock_data.columns = pd.MultiIndex.from_product([['SPY'], df_spy_stock_data.columns])



df_stock_data = pd.concat([df_agg_stock_data, df_spy_stock_data], axis=1)


# Display sample data
df_stock_data.head()
```

```python
# Configuring a Monte Carlo simulation to forecast 30 years cumulative returns
# YOUR CODE HERE!
MC_even_dist = MCSimulation(
    portfolio_data=df_stock_data, 
    weights=[0.4,0.6],
    num_simulation= 500, 
    num_trading_days=252*30
)
```

```python
# Printing the simulation input data
# YOUR CODE HERE!
MC_even_dist.portfolio_data.head()
```

```python
# Running a Monte Carlo simulation to forecast 30 years cumulative returns
# YOUR CODE HERE!
MC_even_dist.calc_cumulative_return()
```

```python
# Plot simulation outcomes
# YOUR CODE HERE!
line_plot = MC_even_dist.plot_simulation()
```

```python
# Plot probability distribution and confidence intervals
# YOUR CODE HERE!
dist_plot = MC_even_dist.plot_distribution()
```

### Retirement Analysis

```python
# Fetch summary statistics from the Monte Carlo simulation results
# YOUR CODE HERE!
even_tbl = MC_even_dist.summarize_cumulative_return()
# Print summary statistics
# YOUR CODE HERE!
print(even_tbl)
```

### Calculate the expected portfolio return at the `95%` lower and upper confidence intervals based on a `$20,000` initial investment.

```python
# Set initial investment
initial_investment = 20000

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $20,000
# YOUR CODE HERE!
ci_lower = round(even_tbl[8]*initial_investment,2)
ci_upper = round(even_tbl[9]*initial_investment,2)
                      
                      
print(f"There is a 95% chance that an initial investment of ${initial_investment} in the portfolio"
      f" over the next 30 years will end within in the range of"
      f" ${ci_lower} and ${ci_upper}")
```

### Calculate the expected portfolio return at the `95%` lower and upper confidence intervals based on a `50%` increase in the initial investment.

```python
# Set initial investment
initial_investment = 20000 * 1.5

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $30,000
# YOUR CODE HERE!

# Print results
print(f"There is a 95% chance that an initial investment of ${initial_investment} in the portfolio"
      f" over the next 30 years will end within in the range of"
      f" ${ci_lower} and ${ci_upper}")
```
